# BEAUTIFUL SOUP

```python
from bs4 import BeautifulSoup
```

```python
s=BeautifulSoup(p12.text,'lxml')
```

```python
print(s.prettify()) #print html en jerarquia
```

```python
secciones=s.find('ul',attrs={'class':'horizontal-list main-sections hide-on-dropdown'}).find_all('li')
```

```python
seccion = secciones[0]
```

```python
seccion.a.get('href') #darle un nombre de un atributo para obtenerlo
seccion.a.get_text() #obtener texto de <a></a>
```

```python
links_secciones = [seccion.a.get('href') for seccion in secciones]
```

```python
links_secciones
```

```python
sec = requests.get(links_secciones[0])
```

```python
s_seccion = BeautifulSoup(sec.text, 'lxml')
```

```python
print(s_seccion.prettify())
```

```python
article_list = s_seccion.find('div',attrs={'class':'articles-list'}).find_all('article')
article_list
```

```python
article = article_list[0]
article_content = article.find('div',attrs={'class':'article-item__content'})
article_content
```

```python
article_content.a.get('href')
article_content.a.get_text()
```

# MANEJO DE ERRORES

```python
#encerrar con try except
url_mala = url.replace('2','3')
try:
    r = requests.get(url_mala)
except Exception as e:
    print(e)
```

# DESCARGANDO CONTENIDO

```python
url_nota= article_content.a.get('href')
url_nota
```

```python
try:
    nota = requests.get(url_nota)
    if(nota.status_code == 200):
        s_nota = BeautifulSoup(nota.text,'lxml')
        #extraemos el titulo
        titulo = s_nota.find('div',attrs={'class':'article-titles'})
        titulo= titulo.find('h1',attrs={'class':'article-title'})
        titulo= titulo.get_text()
        print(titulo)
        #extraer fecha del artículo
        fecha = s_nota.find('span', attrs={'pubdate':'pubdate'}).get('datetime')
        print(fecha)
        #extraer volanta
        volanta = s_nota.find('div',attrs={'class':'article-titles'})
        volanta= volanta.find('h2',attrs={'class':'article-prefix'})
        volanta= volanta.get_text()
        print(volanta)         
except Exception as e:
    print("Error:")
    print(e)
    print('')
```

*Guardar imagen de la nota*

```python
media = s_nota.find('div',attrs={'class':'article-main-media-image'})
```

```python
imagenes = media.find_all('img')
```

```python
if len(imagenes) == 0:
    print('no se encontraron imagenes')
else:
    imagen = imagenes[-1]
    img_src = imagen.get('data-src')
print(img_src)
```

```python
img_req = requests.get(img_src)
```
